```python
def convert_date(date_str):
    try:
        return pd.to_datetime(date_str, format='%m/%d/%Y', errors='coerce').strftime('%m/%d/%Y')
    except ValueError:
        try:
            return pd.to_datetime(date_str, format='%d-%m-%Y', errors='coerce').strftime('%m/%d/%Y')
        except ValueError:
            return date_str 
customers['Birthday'] = customers['Birthday'].apply(convert_date)
```

```python
customers['State Code'] = customers['State Code'].fillna('Not Available')
```

```python
import pandas as pd
import pymysql
from sqlalchemy import create_engine, text

# MySQL connection details
host = "127.0.0.1"
port = "3306"
user = "root"
passwd = "Diya"
database = "gep"

# Step 1: Connect to MySQL server (without specifying a database)
engine_string = f"mysql+pymysql://{user}:{passwd}@{host}:{port}/"
engine = create_engine(engine_string)

# Step 2: Create the database if it doesn't exist
with engine.connect() as conn:
    conn.execute(text(f"CREATE DATABASE IF NOT EXISTS {database}"))
    print(f"Database '{database}' created or already exists.")

# Step 3: Now, connect to the specific database
engine_string = f"mysql+pymysql://{user}:{passwd}@{host}:{port}/{database}"
engine = create_engine(engine_string)


# Step 4: List of DataFrames and their corresponding table names
tables = {
    "overall": overall,
    "customers": customers,
    "products": products,
    "exchangerates": exchangerates,
    "stores": stores
}

# Step 5: Loop through the tables dictionary and import each DataFrame into MySQL
for table_name, df in tables.items():
    try:
        df.to_sql(table_name, engine, if_exists='replace', index=False)
        print(f"Successfully imported {table_name} to SQL")
    except Exception as e:
        print(f"Failed to import {table_name} to SQL: {e}")
```
